### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

```python
# Figure - är själva tavlan eller fönstret där all information/ grafen visas på. Inklusive linjer, prickar, staplar, titlar, ettiketer osv.  
# Axes - grafen, x och y axlarna (z) där själva datan plottas, inkluderar även, graflinejer, ettiketer, skalor och titlar. 
# Axis - Endast en av axlarna i en graf, tex endast x axeln eller endast y axeln. 
# Artist - Alla objekt som kan ritas på en figure, tex linjer, text, figurer och axlar. Primitives (linjer, rektanglar & cirklar), containers(axlar & figurer)
```

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
x = np.linspace(-4,4,100)
y = x**2

# Pyplot API
plt.plot(x,y)

plt.title("y = x^2 (Pyplot figur)")
plt.xlabel("x-axeln")
plt.ylabel("y-axeln")

# Objektorinterad plot
fig, ax = plt.subplots()

ax.plot(x,y)

ax.set_title("y = x^2 Objektorinterad figur")
ax.set_xlabel("x-axeln")
ax.set_ylabel("y-axeln")

```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python
# Skapar figuren med två grafer. 
fig, (ax1, ax2) = plt.subplots(2,1, figsize =(10,10))

# Första grafen, scatter plot
ax1.scatter(random_data_x, random_data_y, alpha = 0.5)
ax1.set_title("SCATTER PLOT")
ax1.set_xlabel("x-axel")
ax1.set_ylabel("y-axeln")

# Andra graden, bar plot
ax2.bar(names, values)
ax2.set_title("BAR PLOT, frukt data")
ax2.set_xlabel("Frukter")
ax2.set_ylabel("Antal")
```
